## Q1 Are passengers of higher socioeconomic status more likely to have survived?
1. Groupby Pclass
2. Calculate mean Survived for each Pclass
3. Test whether or not the means differ significantly from each other.

### Data Cleaning

There are 891 datapoints for 'Survived' and 'Pclass' as expected. 'Pclass' takes values 1, 2 and 3 and has min and max of 1 and 3 as expected. Likewise 'Survived' takes values 0, 1 and has min and max of 0 and 1 as expected.

There is therefore no need for extra data cleaning.

### Analysis

```python
# Group data by passengers' socioeconomic class
q1 = data.groupby('Pclass')
```

```python
# Number of passengers in each socioeconomic class
pclass_count = q1['Pclass'].count()
print(pclass_count)
```

Note that sample size n for each passenger class is large: n > 150 for all Pclass.

```python
# Number of passengers who survived from each socioeconomic class
pclass_survived = q1['Survived'].sum()
print(pclass_survived)
```

```python
# Calculate passenger survival rate for each socioeconomic class
pclass_survived_rate = pclass_survived / pclass_count # Use of vectorized operations
print(pclass_survived_rate)
```

```python
# Scatter plot of results
plt.scatter([1,2,3],pclass_survived_rate)
plt.xlabel('Passenger\'s socioeconomic class')
plt.ylabel('Rate of survival')
```

### Conclusions 

From observation, it seems as though there is a positive correlation between a passenger's socioeconomic class and their rate of survival. Let us test this hypothesis more formally by running an **OLS regression**.
